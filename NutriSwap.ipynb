```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder, StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.metrics.pairwise import cosine_similarity
from sklearn.metrics import classification_report

df = pd.read_csv("/content/drive/MyDrive/indian-food-101/indian_food.csv")
```

```python
## A)  DATA PREPROCESSING
import numpy as np

# 1. Clean the data
df = df.dropna(subset=['ingredients', 'prep_time', 'cook_time'])

# 2. Add more meaningful features
def extract_features(row):
    ingredients_lower = str(row['ingredients']).lower()

    # Healthy ingredients
    healthy_ingredients = ['vegetables', 'greens', 'spinach', 'lentils', 'dal',
                          'chickpea', 'green', 'tomato', 'yogurt', 'curd']
    healthy_score = sum(1 for ing in healthy_ingredients if ing in ingredients_lower)

    # Calculate cooking time score (faster = healthier assumption)
    time_score = 1 / (1 + (row['prep_time'] + row['cook_time']) / 60)

    return pd.Series({
        'healthy_ingredient_count': healthy_score,
        'time_score': time_score
    })

# Apply feature extraction
new_features = df.apply(extract_features, axis=1)
df = pd.concat([df, new_features], axis=1)

print("New features added:")
print(df[['name', 'healthy_ingredient_count', 'time_score']].head())
```

```python
def is_healthy_classifier(ingredients, prep, cook, flavor):
    # Enhanced health classification with multiple factors Returns: 0  or 1 (healthy)
    score = 5  # Start with neutral score
    ingredients_lower = str(ingredients).lower()

    # Penalty for unhealthy ingredients
    unhealthy_words = ["ghee", "oil", "butter", "sugar", "fried", "maida", "cheese", "cream"]
    penalty = sum(1 for word in unhealthy_words if word in ingredients_lower)
    score -= penalty * 0.5

    # Bonus for healthy ingredients
    healthy_words = ["vegetables", "dal", "lentils", "spinach", "green", "yogurt"]
    bonus = sum(1 for word in healthy_words if word in ingredients_lower)
    score += bonus * 0.3

    # Time penalty (longer cooking = more oil absorption)
    if (prep + cook) > 45:
        score -= 1
    elif (prep + cook) < 20:
        score += 0.5

    # Flavor consideration
    if flavor == 'sweet':
        score -= 0.5

    # Return binary classification
    return 1 if score >= 5 else 0

# Apply enhanced classifier
df["healthy"] = df.apply(
    lambda row: is_healthy_classifier(
        row["ingredients"],
        row["prep_time"],
        row["cook_time"],
        row.get("flavor_profile", "")
    ),
    axis=1
)

popularity = df['name'].value_counts().reset_index()
popularity.columns = ['name', 'popularity_score']
df = pd.merge(df, popularity, on='name', how='left')

# Drop the duplicate popularity_score columns after the merge if they exist
df = df.drop(columns=['popularity_score_x', 'popularity_score_y'], errors='ignore')

df_ml = df[["name", "prep_time", "cook_time", "diet", "flavor_profile", "course", "region", "healthy", "popularity_score", "healthy_ingredient_count", "time_score"]].dropna()

le = LabelEncoder()
for col in ["diet", "flavor_profile", "course", "region"]:
    df_ml[col] = le.fit_transform(df_ml[col])
```

```python

#. Text features from ingredients
from sklearn.feature_extraction.text import TfidfVectorizer

# Create TF-IDF features from ingredients (limited to top 20 features)
tfidf = TfidfVectorizer(max_features=20, stop_words='english')
ingredient_features = tfidf.fit_transform(df['ingredients'].fillna(''))
ingredient_df = pd.DataFrame(
    ingredient_features.toarray(),
    columns=[f'ingredient_tfidf_{i}' for i in range(20)]
)

#  Combine with existing features
df_ml = pd.concat([df_ml.reset_index(drop=True), ingredient_df], axis=1)

print("Enhanced feature set shape:", df_ml.shape)
print("\nNew features preview:")
print(df_ml.head())
```

```python
# Prepare features for modeling
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC

# Drop rows with NaN values in y AND track the indices
feature_cols = [col for col in df_ml.columns if col not in ['name', 'healthy']]
X = df_ml[feature_cols]
y = df_ml["healthy"]

nan_mask = y.notna()
X_clean = X[nan_mask]
y_clean = y[nan_mask]

print(f"Original data: {len(df_ml)} rows")
print(f"After removing NaN: {len(y_clean)} rows")
print(f"Removed: {len(df_ml) - len(y_clean)} rows with missing target values\n")

# Scale features
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X_clean)
```

```python
## B) SPLITTING THE DATASET INTO TRAIN,TEST
\
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y_clean, test_size=0.2, random_state=42, stratify=y_clean
)

print("Dataset Split:")
print(f"Training set: {len(X_train)} samples")
print(f"Testing set: {len(X_test)} samples")
print(f"Training healthy dishes: {sum(y_train)}")
print(f"Training unhealthy dishes: {len(y_train) - sum(y_train)}")
print(f"Testing healthy dishes: {sum(y_test)}")
print(f"Testing unhealthy dishes: {len(y_test) - sum(y_test)}")
```

```python
## APPLYING PCA FOR FEATURE REDUCTION

from sklearn.decomposition import PCA
import matplotlib.pyplot as plt

print("="*60)
print("PCA ANALYSIS FOR FEATURE REDUCTION".center(60))
print("="*60)

# Apply PCA to understand feature importance
pca = PCA()
pca.fit(X_scaled)

# Calculate cumulative variance
cumulative_variance = np.cumsum(pca.explained_variance_ratio_)

# Plot explained variance
plt.figure(figsize=(10, 6))
plt.plot(range(1, len(cumulative_variance) + 1), cumulative_variance, 'bo-')
plt.axhline(y=0.95, color='r', linestyle='--', label='95% Variance')
plt.xlabel('Number of Components')
plt.ylabel('Cumulative Explained Variance')
plt.title('PCA: Cumulative Explained Variance')
plt.grid(True)
plt.legend()
plt.show()

# Find optimal number of components for 95% variance
n_components_95 = np.argmax(cumulative_variance >= 0.95) + 1
print(f"\nComponents needed for 95% variance: {n_components_95}/{len(cumulative_variance)}")
print(f"Original features: {X_scaled.shape[1]}")
print(f"Reduced features: {n_components_95}")
print(f"Variance retained: {cumulative_variance[n_components_95-1]:.2%}")

# Apply PCA transformation
pca_optimal = PCA(n_components=n_components_95)
X_pca = pca_optimal.fit_transform(X_scaled)
print(f"\nNew dataset shape: {X_pca.shape}")

# Split PCA-transformed data
X_train_pca, X_test_pca, y_train_pca, y_test_pca = train_test_split(
    X_pca, y_clean, test_size=0.2, random_state=42, stratify=y_clean
)

print(f"PCA Training set: {X_train_pca.shape}")
print(f"PCA Testing set: {X_test_pca.shape}")
```

```python
## C) APPLYING THE ALGORITHM

print("Training models...\n")

# 1. Logistic Regression
lr = LogisticRegression(max_iter=1000, random_state=42)
lr.fit(X_train, y_train)
lr_score = lr.score(X_test, y_test)
print(f"Logistic Regression Test Accuracy: {lr_score:.3f}")

# 2. Support Vector Machine
svm = SVC(probability=True, random_state=42)
svm.fit(X_train, y_train)
svm_score = svm.score(X_test, y_test)
print(f"SVM Test Accuracy: {svm_score:.3f}")

# Select best model
if lr_score >= svm_score:
    best_model = lr
    best_model_name = "Logistic Regression"
    best_score = lr_score
else:
    best_model = svm
    best_model_name = "SVM"
    best_score = svm_score

print(f"\n✅ Best Model: {best_model_name} with accuracy: {best_score:.3f}")
```

```python
## APPLYING KNN CLASSIFIER (BRIEF)

from sklearn.neighbors import KNeighborsClassifier

print("\n" + "="*60)
print("K-NEAREST NEIGHBORS CLASSIFIER".center(60))
print("="*60)

# Find optimal K using cross-validation
from sklearn.model_selection import cross_val_score

k_values = range(3, 16, 2)
k_scores = []

for k in k_values:
    knn = KNeighborsClassifier(n_neighbors=k)
    scores = cross_val_score(knn, X_train, y_train, cv=5, scoring='accuracy')
    k_scores.append(scores.mean())

# Plot K vs Accuracy
plt.figure(figsize=(8, 5))
plt.plot(k_values, k_scores, 'bo-', linewidth=2, markersize=8)
plt.xlabel('K Value')
plt.ylabel('Cross-Validation Accuracy')
plt.title('KNN: Finding Optimal K')
plt.grid(True)
plt.show()

# Train with optimal K
optimal_k = k_values[np.argmax(k_scores)]
knn = KNeighborsClassifier(n_neighbors=optimal_k)
knn.fit(X_train, y_train)
knn_score = knn.score(X_test, y_test)

print(f"\nOptimal K: {optimal_k}")
print(f"KNN Test Accuracy: {knn_score:.3f}")

# Compare with existing models
print(f"\n Model Comparison:")
print(f"Logistic Regression: {lr_score:.3f}")
print(f"SVM: {svm_score:.3f}")
print(f"KNN: {knn_score:.3f}")
```

```python
## APPLYING MLP NEURAL NETWORK (MINIMAL)

from sklearn.neural_network import MLPClassifier
from sklearn.metrics import classification_report

print("\n" + "="*60)
print("MULTI-LAYER PERCEPTRON (NEURAL NETWORK)".center(60))
print("="*60)

# Train MLP with minimal architecture
mlp = MLPClassifier(
    hidden_layer_sizes=(50, 25),  # 2 hidden layers
    activation='relu',
    solver='adam',
    max_iter=500,
    random_state=42,
    early_stopping=True,
    validation_fraction=0.1
)

print("Training MLP Neural Network...")
mlp.fit(X_train, y_train)

# Evaluate
mlp_score = mlp.score(X_test, y_test)
mlp_pred = mlp.predict(X_test)

print(f"\n MLP Test Accuracy: {mlp_score:.3f}")
print(f"\n Classification Report:")
print(classification_report(y_test, mlp_pred, target_names=['Unhealthy', 'Healthy']))

# Plot training loss
plt.figure(figsize=(8, 5))
plt.plot(mlp.loss_curve_, linewidth=2)
plt.xlabel('Iteration')
plt.ylabel('Loss')
plt.title('MLP Training Loss')
plt.grid(True)
plt.show()

# Final comparison
print("\n" + "="*60)
print("FINAL MODEL COMPARISON".center(60))
print("="*60)
print(f"Logistic Regression: {lr_score:.3f}")
print(f"SVM: {svm_score:.3f}")
print(f"KNN: {knn_score:.3f}")
print(f"MLP Neural Network: {mlp_score:.3f}")

# Select best overall model
all_scores = {
    'Logistic Regression': lr_score,
    'SVM': svm_score,
    'KNN': knn_score,
    'MLP': mlp_score
}

best_model_name = max(all_scores, key=all_scores.get)
print(f"\n Best Overall Model: {best_model_name} ({all_scores[best_model_name]:.3f})")
```

```python
from colorama import Fore, Back, Style, init
init(autoreset=True)

def recommend_healthy_snack(dish_name, top_n=3, show_reasons=True):
    """Enhanced recommendation system with beautiful output"""

    # Verify health_score exists in df
    if 'health_score' not in df.columns:
        print("❌ Error: health_score not found in main dataframe!")
        print("Please run the merge cell first.")
        return

    # Find user's dish (case-insensitive)
    user_dish_row = df[df['name'].str.lower() == dish_name.lower()]

    if user_dish_row.empty:
        print(f"❌ Sorry, '{dish_name}' is not in our database.")
        print("\n📋 Here are some available dishes:")
        sample_dishes = df['name'].sample(min(10, len(df))).tolist()
        for i, dish in enumerate(sample_dishes, 1):
            print(f"   {i}. {dish}")
        return

    # Get user dish information
    orig_dish = user_dish_row.iloc[0]
    user_health_score = orig_dish['health_score']

    # ========== ENHANCED HEADER ==========
    print("\n" + Fore.CYAN + "═" * 85)
    print(Fore.CYAN + Style.BRIGHT + "🌿 NUTRISWAP 🌿".center(85))
    print(Fore.GREEN + "AI-Powered Nutritional Analysis & Recommendations".center(85))
    print(Fore.CYAN + "═" * 85 + Style.RESET_ALL)

    # ========== USER SELECTION BOX ==========
    print("\n" + Fore.MAGENTA + "╔" + "═" * 83 + "╗")
    print(Fore.MAGENTA + "║ " + Fore.WHITE + Style.BRIGHT +
          f"📋 YOUR SELECTION: {orig_dish['name'].upper()}".ljust(84) +
          Fore.MAGENTA + "║")
    print(Fore.MAGENTA + "╠" + "═" * 83 + "╣")

    # Health score with color coding
    if user_health_score < 0.4:
        score_color = Fore.RED + "⚠️  Low Health"
    elif user_health_score < 0.7:
        score_color = Fore.YELLOW + "⚡ Moderate Health"
    else:
        score_color = Fore.GREEN + "✅ High Health"

    print(Fore.MAGENTA + "║ " + Fore.WHITE +
          f"Health Score: {score_color} ({user_health_score:.2f})".ljust(93) +
          Fore.MAGENTA + "║")

    region_display = orig_dish['region'] if orig_dish['region'] not in ['', '-1', -1] else 'Common'
    print(Fore.MAGENTA + "║ " + Fore.WHITE +
          f"Diet: {Fore.CYAN}{orig_dish['diet'].title()}  {Fore.WHITE}│  Region: {Fore.CYAN}{region_display}".ljust(93) +
          Fore.MAGENTA + "║")
    print(Fore.MAGENTA + "║ " + Fore.WHITE +
          f"Preparation: {Fore.CYAN}{orig_dish['prep_time']} min prep • {orig_dish['cook_time']} min cook".ljust(93) +
          Fore.MAGENTA + "║")
    print(Fore.MAGENTA + "╚" + "═" * 83 + "╝" + Style.RESET_ALL)

    # Calculate similarity based on features
    try:
        dish_in_ml = df_ml[df_ml['name'].str.lower() == dish_name.lower()]

        if not dish_in_ml.empty and dish_in_ml.index[0] in X_clean.index:
            clean_position = X_clean.index.get_loc(dish_in_ml.index[0])
            user_features = X_scaled[clean_position].reshape(1, -1)
            similarities_clean = cosine_similarity(user_features, X_scaled).flatten()
            similarity_dict = dict(zip(X_clean.index, similarities_clean))
            df['similarity'] = df_ml.index.map(similarity_dict).fillna(0.3)
            use_similarity = True
        else:
            df['similarity'] = 0.5
            use_similarity = False
    except Exception as e:
        df['similarity'] = 0.5
        use_similarity = False

    # Filter for healthier alternatives
    healthier_mask = (
        (df['health_score'] > user_health_score) &
        (df['similarity'] > 0.25) &
        (df['name'].str.lower() != dish_name.lower())
    )

    recommendations_df = df[healthier_mask].copy()

    if recommendations_df.empty:
        print(f"\n" + Fore.GREEN + Style.BRIGHT + "✅ '{orig_dish['name']}' is already quite healthy!" + Style.RESET_ALL)
        print("\n💡 Other healthy options from the same region:")

        region_healthy = df[
            (df['region'] == orig_dish['region']) &
            (df['health_score'] > 0.6) &
            (df['name'].str.lower() != dish_name.lower())
        ].sort_values('health_score', ascending=False).head(3)

        if not region_healthy.empty:
            for _, row in region_healthy.iterrows():
                print(f"   • {row['name']} (Score: {row['health_score']:.2f})")
        return

    # Calculate final score
    max_pop = df['popularity_score'].max() if df['popularity_score'].max() > 0 else 1

    recommendations_df['final_score'] = (
        0.5 * recommendations_df['health_score'] +
        0.3 * recommendations_df['similarity'] +
        0.2 * (recommendations_df['popularity_score'] / max_pop)
    )

    # Get top recommendations
    top_recs = recommendations_df.sort_values('final_score', ascending=False).head(top_n)

    # ========== RECOMMENDATIONS HEADER ==========
    print("\n" + Fore.GREEN + Style.BRIGHT + "╔" + "═" * 83 + "╗")
    print(Fore.GREEN + "║ " + Fore.WHITE + Style.BRIGHT +
          "🌟 HEALTHIER ALTERNATIVES - AI RECOMMENDED".center(84) +
          Fore.GREEN + "║")
    print(Fore.GREEN + "╚" + "═" * 83 + "╝" + Style.RESET_ALL)

    # ========== RECOMMENDATIONS ==========
    for idx, (_, row) in enumerate(top_recs.iterrows(), 1):
        improvement = row['health_score'] - user_health_score

        # Choose icon based on improvement
        if improvement > 0.5:
            icon = "🔥"
            improvement_color = Fore.GREEN + Style.BRIGHT
        elif improvement > 0.3:
            icon = "⚡"
            improvement_color = Fore.YELLOW
        else:
            icon = "✨"
            improvement_color = Fore.CYAN

        print("\n" + Fore.BLUE + "┌" + "─" * 83 + "┐")
        print(Fore.BLUE + "│ " + Fore.WHITE + Style.BRIGHT +
              f"#{idx}  {row['name'].upper()}".ljust(84) + Fore.BLUE + "│")
        print(Fore.BLUE + "├" + "─" * 83 + "┤")

        print(Fore.BLUE + "│  " + Fore.WHITE +
              f"Health Score: {Fore.GREEN}{row['health_score']:.2f} " +
              improvement_color + f"{icon} +{improvement:.2f} improvement".ljust(73) +
              Fore.BLUE + "│")

        rec_region = row['region'] if row['region'] not in ['', '-1', -1] else 'Common'
        print(Fore.BLUE + "│  " + Fore.WHITE +
              f"Similarity: {Fore.CYAN}{row['similarity']:.0%}  {Fore.WHITE}│  " +
              f"Diet: {Fore.YELLOW}{row['diet'].title()}  {Fore.WHITE}│  " +
              f"Region: {Fore.MAGENTA}{rec_region}".ljust(73) + Fore.BLUE + "│")

        # Time display
        try:
            total_time = float(row['prep_time']) + float(row['cook_time'])
            if total_time > 0 and total_time < 30:
                time_display = f"{Fore.GREEN}⚡ Quick prep ({int(total_time)} min)"
            elif total_time > 0:
                time_display = f"{Fore.YELLOW}📅 {int(total_time)} min total"
            else:
                time_display = f"{Fore.CYAN}Quick prep"
        except (ValueError, TypeError):
            time_display = f"{Fore.CYAN}Quick prep"

        print(Fore.BLUE + "│  " + time_display + "".ljust(73) + Fore.BLUE + "│")

        # Show reasons if enabled
        if show_reasons:
            reasons = []
            try:
                orig_time = float(orig_dish['prep_time']) + float(orig_dish['cook_time'])
                row_time = float(row['prep_time']) + float(row['cook_time'])
                if orig_time > 0 and row_time > 0 and row_time < orig_time * 0.8:
                    reasons.append(f"Faster to prepare")
            except (ValueError, TypeError):
                pass

            if row['region'] == orig_dish['region'] and row['region'] not in ['', '-1', -1]:
                reasons.append("Same region preference")

            if reasons:
                print(Fore.BLUE + "│  " + Fore.CYAN + "✨ " + ", ".join(reasons) +
                      "".ljust(73) + Fore.BLUE + "│")

        print(Fore.BLUE + "└" + "─" * 83 + "┘" + Style.RESET_ALL)

    # ========== FOOTER ==========
    print("\n" + Fore.CYAN + "═" * 85)
    print(Fore.YELLOW + "💡 PROJECT GOAL: ".ljust(20) + Fore.WHITE +
          "Promote healthier eating through intelligent food recommendations")
    print(Fore.CYAN + "═" * 85 + Style.RESET_ALL + "\n")
```
